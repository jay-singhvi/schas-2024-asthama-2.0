```python
import pandas as pd
import matplotlib.pyplot as plt

data_file = '/Users/sdayama/Desktop/Population_Analysis/schas-2024-asthama-2.0/Individual_EDA/patient_statistics.csv'
data = pd.read_csv(data_file)
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

axs[0].boxplot([data['M_mean'], data['A_mean']], labels=['Morning Mean', 'Afternoon Mean'])
axs[0].set_title('Mean Distribution')

axs[1].boxplot([data['M_median'], data['A_median']], labels=['Morning Median', 'Afternoon Median'])
axs[1].set_title('Median Distribution')

axs[2].boxplot([data['M_std'], data['A_std']], labels=['Morning Std', 'Afternoon Std'])
axs[2].set_title('Standard Deviation Distribution')

plt.tight_layout()
plt.show()
average_morning = data['M_mean'].mean()
average_afternoon = data['A_mean'].mean()

std_morning = data['M_std'].mean()
std_afternoon = data['A_std'].mean()

print(f"Average Morning Mean: {average_morning}")
print(f"Average Afternoon Mean: {average_afternoon}")
print(f"Average Standard Deviation Morning : {std_morning}")
print(f"Average Standard Deviation Afternoon : {std_afternoon}")
```
