```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read the CSV file
df = pd.read_csv("data/translated.csv")

# Basic statistics
display(df.describe())
```

```python
# Age distribution
plt.figure(figsize=(10, 6))
sns.histplot(df["Age"], bins=20, kde=True)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.savefig("Visual_EDA/age_distribution.png")
plt.show()
plt.close()

# BMI distribution
plt.figure(figsize=(10, 6))
sns.histplot(df["BMI"], bins=20, kde=True)
plt.title("BMI Distribution")
plt.xlabel("BMI")
plt.ylabel("Count")
plt.savefig("Visual_EDA/bmi_distribution.png")
plt.show()
plt.close()

# Smoking status
smoking_counts = df["Smoke"].value_counts()
plt.figure(figsize=(8, 6))
smoking_counts.plot(kind="pie", autopct="%1.1f%%")
plt.title("Smoking Status Distribution")
plt.ylabel("")
plt.savefig("Visual_EDA/smoking_status.png")
plt.show()
plt.close()

# Occupation distribution
occupation_counts = df["occupation_category"].value_counts()
plt.figure(figsize=(12, 6))
occupation_counts.plot(kind="bar")
plt.title("Occupation Distribution")
plt.xlabel("Occupation Category")
plt.ylabel("Count")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.savefig("Visual_EDA/occupation_distribution.png")
plt.show()
plt.close()

# Correlation heatmap
correlation_matrix = df[
    ["Age", "Height", "Weight", "BMI", "BSA", "M_mean", "A_mean"]
].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.tight_layout()
plt.savefig("Visual_EDA/correlation_heatmap.png")
plt.show()
plt.close()

# Age vs BMI scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x="Age", y="BMI", hue="Sex")
plt.title("Age vs BMI")
plt.savefig("Visual_EDA/age_vs_bmi.png")
plt.show()
plt.close()

print("Analysis complete. Check the generated PNG files for visualizations.")
```
